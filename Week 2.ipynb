```python
student = {
    "name": "Ahmad",
    "age": 25,
    "gpa": 3.8
}

# Method 1: dict["key"] - direct access
print(student["name"]) # Ahmad
print(student["gpa"]) # 3.8

# Method 2: .get() - safe access (no crash)
print(student.get("age")) # 25
print(student.get("email")) # None
print(student.get("email", "N/A")) # N/A
```

```python
import requests

# 1. Define the real API URL from The Edge Malaysia
url = "https://theedgemalaysia.com/api/loadMoreCategories?offset=0&categories=malaysia"

# 2. Ask the website for the data and automatically convert the JSON into a Python dictionary
response = requests.get(url)
data = response.json()

# --- HOW TO ACCESS THE NESTED DATA ---

# A. Access a top-level key (Returns a number)
total_articles = data["total"]
print(f"Total articles available: {total_articles}")
print("-" * 30)

# B. Access the first article's title 
# data is a dict -> "results" is a list -> [0] is the first item (a dict) -> ["title"] is the key
first_title = data["results"][0]["title"]
print(f"First Headline: {first_title}")
print("-" * 30)

# C. Loop through every article in the list and print their titles
print("All Headlines on this page:")
for article in data["results"]:
    print(f"- {article['title']}")
```

```python
# The nested dictionary setup from your image
countries = {
    'Malaysia': {
        'capital': 'Kuala Lumpur',
        'population': 32000000,
        'currency': 'Malaysian Ringgit'
    },
    'Singapore': {
        'capital': 'Singapore',
        'population': 5600000,
        'currency': 'Singapore Dollar'
    }
}

# 1. Print Malaysia's full data
print(countries['Malaysia'])

# 2. Print the capital of Malaysia
print(countries['Malaysia']['capital'])

# 3. Print the population of Singapore
print(countries['Singapore']['population'])

# 4. Print the total population of Malaysia + Singapore
total_population = countries['Malaysia']['population'] + countries['Singapore']['population']
print(total_population)
```
