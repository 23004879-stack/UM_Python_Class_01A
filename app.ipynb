```python
buy_price = float(input('Enter your buy price: '))
stock_name = 'Maybank'
sell_price = float(input('Enter your sell price: '))

# 1. Calculate the profit and save it to a variable first
profit = sell_price - buy_price
return_percentage = (profit / buy_price) * 100

# 2. Print outputs (.upper() makes 'Maybank' ALL CAPS)
print(f"Stock: {stock_name.upper()}")

# :.2f forces Python to display exactly 2 decimal places
print(f"Buy price:  RM {buy_price:.2f}")
print(f"Sell price: RM {sell_price:.2f}")

# 3. Add the divider line shown in the expected output
print("___________") 
print() # Adds an empty space line

print(f"Profit/Loss: RM {profit:.2f}")
print(f"Return: {return_percentage:.2f}%")
```
